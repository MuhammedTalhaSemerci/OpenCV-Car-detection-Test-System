```python
import cv2
import numpy as np
import time
import torch
net = cv2.dnn.readNet('yolov3.weights','yolov3.cfg')
classes = []

with open ('coco.names','r') as f:
    classes=f.read().splitlines()


    
    
cap = cv2.VideoCapture(0)



cnt = 0

sayi = 0


w=cap.get(3)
h=cap.get(4)








```

```python

#img= cv2.imread('index.jpeg')

while True:
    
    epoch = time.localtime()
    _,img = cap.read()
    height,width, _ = img.shape

    blob = cv2.dnn.blobFromImage(img,1/255,(416,416),(0,0,0),swapRB=True,crop=False)

    net.setInput(blob)

    output_layers_names = net.getUnconnectedOutLayersNames()
    layersOutputs = net.forward(output_layers_names)

    
    
    boxes = []
    confidences = []
    class_ids = []
    
    cv2.line(img,(600,250),(600,650),(255,0,0),10)
    
    for output in layersOutputs:
         for detection in output:
                scores = detection [5:]
                class_id = np.argmax(scores)
                confidence = scores[class_id]
                if confidence > 0.5:
                    center_x = int(detection[0]*width)
                    center_y = int(detection[1]*height)
                    w = int(detection[2]*width)
                    h = int(detection[3]*height)                

                    x = int(center_x - w/2)
                    y = int(center_y - h/2)

                    boxes.append([x,y,w,h])
                    confidences.append(float(confidence))

                    class_ids.append(class_id)

    indexes = cv2.dnn.NMSBoxes(boxes,confidences,0.5,0.4)

    font = cv2.FONT_HERSHEY_PLAIN
    colors = np.random.uniform(0,255,size=(len(boxes),3) )

    if len(indexes)>0:
        for i in indexes.flatten():
            x,y,w,h = boxes[i]
            label = str(classes[class_ids[i]])
           
            
        
            confidence = str(round(confidences[i],2))
            color = colors[i]
            cv2.rectangle(img,(x,y),(x+w,y+h),color,2)
            cv2.putText(img,label + " " +confidence , (x,y+20),font,2,(255,255,255) ,2)

            line_cx = 600
            car_cx = int(x+w/2)
            
            if (epoch.tm_hour >= 13 and epoch.tm_min >= 0 and epoch.tm_sec >= 0): 
                
                
                if label in ["car"]:

                    if (car_cx < line_cx+300 and car_cx > line_cx-300):

                        cnt = cnt+1
                        print(cnt)
         
        
        
        
            #ilk aşama
                    
    if (epoch.tm_hour == 13 and epoch.tm_min == 15 and epoch.tm_sec == 0 and sayi == 0):
        file = open('yazikayit_deneme.txt', 'a') 
        
        file.write("12.)")
        file.write(str(cnt))
        file.write("\n")
        file.close()     
        sayi = 1
    
    if (epoch.tm_hour == 13 and epoch.tm_min == 15 and epoch.tm_sec == 1 and sayi == 1):               
        
        sayi = 0
        cnt = 0
              
    
    
    
    
          #ikinci aşama
    
    
    if (epoch.tm_hour == 13 and epoch.tm_min == 30 and epoch.tm_sec == 0 and sayi == 0):
        file = open('yazikayit_deneme.txt', 'a') 
        
        file.write("15.)")
        file.write(str(cnt))
        file.write("\n")
        file.close()  
        
        sayi = 1
    
    if (epoch.tm_hour == 13 and epoch.tm_min == 30 and epoch.tm_sec == 1 and sayi == 1):               
        
        sayi = 0
        cnt = 0
    
    
    
    
    
    
          #üçüncü aşama
    
    if (epoch.tm_hour == 13 and epoch.tm_min == 45 and epoch.tm_sec == 0 and sayi == 0):
        file = open('yazikayit_deneme.txt', 'a') 
        
        file.write("18.)")
        file.write(str(cnt))
        file.write("\n")
        file.close()     
        sayi = 1
    
    if (epoch.tm_hour == 13 and epoch.tm_min == 45 and epoch.tm_sec == 1 and sayi == 1):               
        
        sayi = 0
        cnt = 0
    
           
        
        
        
        
          #dördüncü aşama
    
    
    if (epoch.tm_hour == 14 and epoch.tm_min == 0 and epoch.tm_sec == 0 and sayi == 0):
        file = open('yazikayit_deneme.txt', 'a') 
        
        file.write("21.)")
        file.write(str(cnt))
        file.write("\n")
        file.close()     
        sayi = 1
    
    if (epoch.tm_hour == 14 and epoch.tm_min == 0 and epoch.tm_sec == 1 and sayi == 1):               
        
        sayi = 0
        cnt = 0
    
    
    
                                        
    cv2.imshow('Video',img)
    key = cv2.waitKey(1)
    
    if key == 27:
        break
   
   
cap.release()
cv2.destroyAllWindows()
```
